```python
import pandas as pd
df1=pd.DataFrame({
"Name":["Priyanka","Manan","Summit","Suhani"],
"id":[1,2,3,4]
})

df2=pd.DataFrame({
"Course":["MBA","M.Tech","CAT","CA","NEET","REET"],
"id":[1,2,3,4,5,6],

})
res = df1.merge(df2, how="left", on="id")
print(res)
print()
res = df1.merge(df2, how="right", on="id")
print(res)
```

```python
#index based join
import pandas as pd
df1=pd.DataFrame({
"Name":["Priyanka","Manan","Summit","Suhani"],
"Marks":[70,80,77,66]
},
index=[1,2,3,4])

df2=pd.DataFrame({
"Course":["MBA","M.Tech","CAT","CA"],
"Marks":[53,64,71,70]

},
index=[1,2,3,4])

res=df1.join(df2,lsuffix="df1",rsuffix="df2")
print(res)
```

```python
#right join on common column using pd.merge

import pandas as pd
df1 = pd.DataFrame({
    "Name": ["Priyanka", "Manan", "Summit", "Suhani","Minti"],
    "Year": [2023, 2024, 2025, 2026,2027],
    "id": [1, 2, 3, 4,5]
})

df2 = pd.DataFrame({
"Course":["MBA","M.Tech","CAT","CA"],
 "Year": [2023, 2024, 2025,2026],
"id":[1,2,3,4],

})
res = pd.merge(df1, df2, how="right", on=["id"])
print(res)
print()
res = pd.merge(df1, df2, how="right", on=["id","Year"])
print(res)
```

```python
import pandas as pd
df1 = pd.DataFrame({
    "Name": ["Priyanka", "Manan", "Summit", "Suhani","Minti"],
    "Year": [2023, 2024, 2025, 2026,2027],
    "id": [1, 2, 3, 4,5]
})

df2 = pd.DataFrame({
"Name":["Kartik","Drishti","Dipti","Dhruvi"],
 "Year": [2023, 2024, 2025,2026],
"id":[1,2,3,4],

})
df3 = pd.DataFrame({
    "Name": ["arpita", "Manvi", "kriti", "Supriya"],
    "Year": [2000, 2001, 2002, 2006],
    "id": [1, 2, 3, 4]
})

#Vertical concatenation
res = pd.concat([df1, df2], ignore_index=True)
print(res)
print()
s= res.merge(df3,on="id")
print(s)
```

```python
print("diff between pd.merge() and df.join():")
print("""
      ---pd.merge()---
1) pd.merge() performs column-based joining, i.e., it joins DataFrames using common columns.
2) syntax of pd.merge :"pd.merge(df1, df2, on="ID", how="inner")   
3) pd.merge() is more flexible:
>Can join on columns
>Supports multiple keys
>Can use different column names (left_on, right_on)

  ---df.join()---  
1)df.join() performs index-based joining, i.e., it joins DataFrames using index values.
2)syntax of df.join():left.join(right)
3)df.join() is less flexible:
>Mainly works with index-based joining
>Limited support for complex joins""")
```
