```python
# PROGRAM TO WRITE A CSV FILE 
import csv
with open("address_book.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Address", "Mobile", "Email"])

    n = int(input("Enter number of records: "))

    for i in range(n):
        print(f"\nEnter details for Record {i+1} :")
        name = input("Enter Name: ")
        address = input("Enter Address: ")
        mobile = input("Enter Mobile Number: ")
        email = input("Enter Email: ")
        writer.writerow([name, address, mobile, email])

print("\nAddress Book CSV file created successfully!")

with open("address_book.csv", "r") as file:
    reader = csv.reader(file)
    print("\nContents of the Address Book CSV file:")
    for row in reader:
        print(row)
```

```python
import sqlite3
conn = sqlite3.connect("college.db")
cursor = conn.cursor()
print("Database created successfully!")
```

```python
# Student Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student(
    student_id INTEGER PRIMARY KEY,
    name VARCHAR(100),
    class INTEGER,
    marks FLOAT(5, 2)
)
""")

# Teacher Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Teacher(
    teacher_id INTEGER PRIMARY KEY,
    name VARCHAR(100),
    subject VARCHAR(50)
)
""")

# Course Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Course(
    course_id INTEGER PRIMARY KEY,
    course_name VARCHAR(100)
)
""")

print("Tables created successfully!")
```

```python
# Insert into Student table
cursor.execute("INSERT INTO Student VALUES (1, 'Sakshi', 8, 84.00)")
cursor.execute("INSERT INTO Student VALUES (2, 'Preeti', 11, 77.00)")
cursor.execute("INSERT INTO Student VALUES (3, 'Priya', 12, 92.00)")

# Insert into Teacher table
cursor.execute("INSERT INTO Teacher VALUES (1, 'Mukund Sir', 'Python')")
cursor.execute("INSERT INTO Teacher VALUES (2, 'Mishra Mam', 'DBMS')")

# Insert into Course table
cursor.execute("INSERT INTO Course VALUES (1, 'Python Programming')")
cursor.execute("INSERT INTO Course VALUES (2, 'Database Management')")

conn.commit()

print("Records inserted successfully!")
```

```python
print("All Students:")
cursor.execute("SELECT * FROM Student")
for row in cursor.fetchall():
    print(row)

print("\nStudents with Marks > 80")
cursor.execute("SELECT * FROM Student WHERE marks > 80")
for row in cursor.fetchall():
    print(row)

print("\nTeacher Names:")
cursor.execute("SELECT DISTINCT name FROM Teacher")
for row in cursor.fetchall():
    print(row)

print("\nCourses in descending order of names:")
cursor.execute("SELECT * FROM Course ORDER BY course_name DESC")
for row in cursor.fetchall():
    print(row)
```

```python
cursor.execute("""
UPDATE Student
SET marks = 82
WHERE student_id = 2
""")

conn.commit()

print("Student data updated successfully!")

print("\nUpdated Student Record:")
cursor.execute("SELECT * FROM Student WHERE student_id = 2")
print(cursor.fetchone())
```

```python
cursor.execute("""
DELETE FROM Teacher
WHERE teacher_id = 2
""")
conn.commit()
print("Teacher record deleted successfully!")

print("\nRemaining Teachers:")
cursor.execute("SELECT * FROM Teacher")
for row in cursor:
    print(row)

conn.close()
print("\nDatabase connection closed!")
```
