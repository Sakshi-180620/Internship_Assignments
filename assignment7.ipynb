```python
#ques3
import pandas as pd

df = pd.DataFrame({
    "Name": ["Anuj", "Kriti", "kishan"],
    "Age": [20, 21, 22]
})

for i in range(len(df)):
    print(df.iloc[i]["Name"], df.iloc[i]["Age"])
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Anika", "Riya", "Priya", "Neha"],
    "Marks": [80, 88, 75, 90]
})

# Drop rows where Marks are greater than 80
df = df[df["Marks"] >= 80]

print(df)
```
