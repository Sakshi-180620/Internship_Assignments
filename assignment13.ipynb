```python
# QUES 3
import pandas as pd
import numpy as np
data = {
    "Math": [60, 90, np.nan, 56],
    "Science": [85, np.nan, 55, 85]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df.fillna(df.mean(),inplace=True)
print("\nDataFrame after Replacing NaN with Column Average:")
print(df)
```

```python
import numpy as np
arr=np.array([[10,20,11],[10,5,5]])
arr1=np.array([[10,20,4],[10,2,5]])
print("First 2-D Array:")
print(arr)
print()
print("Second 2-D Array:")
print(arr1)
print()
avg=np.average(arr1+arr/2)
print("Average of the two 2-D arrays:", avg )
print()

from statistics import mode
combined = np.concatenate((arr1, arr))

print("Mean:", np.mean(combined))
print("Median:", np.median(combined))
print("Mode:", mode(combined.flatten()))
```

```python
# QUES 6
A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])
B = np.array([9, -6, 17])

A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("Solution:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

```python
# QUES 7
import matplotlib.pyplot as plt
subjects = ["Math", "Python", "DBMS", "OS", "DSA"]

sem1 = [75, 80, 70, 75, 72]
sem2 = [85, 88, 80, 70, 83]

plt.plot(subjects, sem1,
         marker='o',
         linewidth=2,
         label='Semester 1')
plt.plot(subjects, sem2,
         marker='s',
         linewidth=2,
         label='Semester 2')

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.show()
```
