```python
#1. Function to find maximum of three numbers

def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Taking input from user
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Maximum number is:", maximum(x, y, z))
```

```python
#2. Function to return distinct elements from a list
def dist_fun(lst):
    new_lst=[]
    for i in lst:
        if i not in new_lst:
            new_lst.append(i)
    return new_lst
l=[10,2,2,3,6,8]
print('original lst:',l)
print('new lst:',dist_fun(l))


```

```python
#3. Function to multiply all numbers in a list
def multiply_lst(lst):
    result=1
    for i in lst:
        result = result * i
    return result
num=[2,3,8,10,6]
print('multiplication :',multiply_lst(num))
```

```python
#4. Function to calculate factorial
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
n=int(input('enter a number:'))
print('factorial of a number', n,'is:',factorial(n))
```

```python
#5. Function to reverse a string
def reverse_string(s):
    return s[::-1]
string=input('enter a string:')
print('reverse of string',string,'is:', reverse_string(string))



```

```python
#6. Function to check whether a number falls within a given range
def check_range(num,start,end):
    if num>=start and num<=end:
        print(num,'is in the range')
    else:
        print(num,'is not in the range')
num=int(input('enter a number:'))
start=int(input('enter starting number:'))
end=int(input('enter ending number:'))
check_range(num,start,end)
```

```python
#7. Function to print even numbers from a given list
def even_num(lst):
    even_lst=[]
    for i in lst:
        if i%2==0:
            even_lst.append(i)
    return even_lst
num=[1,2,8,9,0,4,6,]
print('even numbers in the list:',even_num(num))
```

```python
#8. Function to check whether a number is prime or not
def prime_num(n):
    if n <=1:
        print('not prime')
    else:
     for i in range(2,n):
      if n%i==0:
       print('not prime')
      break
     else:
        print('prime')
number=int(input('enter a number:'))
print('number is:',number)
prime_num(number)
```

```python
#9. Function to count the number of uppercase and lowercase letters in a string
def count_letters(text):

    upper = 0
    lower = 0

    for i in text:

        if i.isupper():
            upper = upper + 1

        elif i.islower():
            lower = lower + 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

string = input("Enter a string: ")
print('string is',string)

count_letters(string)
```
