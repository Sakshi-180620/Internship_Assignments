```python
name = input('enter student name:')
stu_class = input('enter student class:')
total = 0
for i in range(1,6):
    marks = float(input(f'enter marks of subject {i}: '))
    total = total + marks
    percentage = (total/500)*100
print('student name:',name)
print('student class:',stu_class)
print('total marks:',total)
print('percentage:',percentage)
 
```

```python
str='learning'
str1='Python'
str2='1234'
print('concatenated string:', str + ' ' + str1) 
print('lower case:', str.lower())
print('upper case:', str.upper())
print('capitalized:', str1.capitalize())
print('after replacement:', str1.replace('o','y'))
print('casefolded:', str1.casefold())
print('centered:', str.center(50))
print('count of "n":', str.count('n'))
print('index of "P":', str1.find('P'))
print('isalpha:', str1.isalpha())
print('isdigit:', str2.isdigit())
print('isnumeric:', str2.isnumeric())
print('endswith "s":', str.endswith('s'))
print('isspace:', str1.isspace())
print('swapcase:', str1.swapcase())
print('title:', str1.title())
```

```python
# assignment operators
a = 50

print("Initial value:", a)

a += 5
print("After += :", a)

a -= 3
print("After -= :", a)

a *= 2
print("After *= :", a)

a /= 4
print("After /= :", a)

a %= 3
print("After %= :", a)

a //= 2
print("After //= :", a)

a **= 3
print("After **= :", a)
```

```python
name = input('enter student name:')
stu_class = input('enter student class:')
total = 0
for i in range(1,6):
    marks = float(input(f'enter marks of subject {i}: '))
    total = total + marks
    percentage = (total/500)*100
print('student name:',name)
print('student class:',stu_class)
print('total marks:',total)
print('percentage:',percentage)
if percentage >=60:
    print('grade A')
elif percentage >=50 and percentage <60:
    print('grade B')
elif percentage >=40 and percentage <50:
    print('grade C')
elif percentage >=30 and percentage <40:
    print('grade D')
else:
    print('Fail')
```
